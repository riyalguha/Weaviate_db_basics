```python
#Delete the schema if it alredy exists
if client.schema.exists("Question"):
    client.schema.delete_class("Question")
```

```python
#Create the schema that will house our data
class_obj = {
    "class": "Question",
    "vectorizer": "text2vec-openai",
}
```

```python
#Use the object above to create the schema
client.schema.create_class(class_obj)
```

```python
with client.batch.configure() as batch:
    # for i, d in enumerate(data):  # Batch import data

        print(f"importing question: {10}")

        #Specify the properties we want to import into Weviate

        properties = {
            "answer": data[9]["Answer"],
            "question":data[9]["Question"],
            'category':data[9]["Category"]
        }

        #Add data to Weaviate

        batch.add_data_object(
            data_object=properties,
            class_name="Question")
```

```python
json_print(client.query.aggregate("Question").with_meta_count().do())
```

```python
#Extract and show any 3 questions and answers

json_print(client.query.get("Question", ["question","answer"]).with_limit(3).do())
```

```python
# write a query to extract the vector for a question

result = (client.query
          .get("Question", ["category", "question", "answer"])
          .with_additional("vector")
          .with_limit(1)
          .do())
```

```python
#Build a vector search query to extract questions ,answers and categories related to "biology"

response = (
    client.query
    .get("Question",["question","answer","category"])
    .with_near_text({"concepts":["biology"]})
    .with_limit(3)
    .do()
)

print(json.dumps(response, indent=4))
```

```python
#Write code to extract the distance between the query and returned object vectors

response = (
    client.query
    .get("Question",["question","answer","category"])
    .with_near_text({"concepts":["biology"]})
    .with_additional('distance')
    .with_limit(2)
    .do()
)

print(json.dumps(response, indent=4))
```

```python
#Extract all 10 questions and analyze them based on distance/similarity to the query vector

response = (
    client.query
    .get("Question", ["question", "answer"])
    .with_near_text({"concepts": ["animals"]})
    .with_limit(10)
    .with_additional(["distance"])
    .do()
)

print(json.dumps(response, indent=2))
```

```python
#Set a max distance threshold - What should the max distance be?
max_distance = 0.24

response = (
    client.query
    .get("Question", ["question", "answer"])
    .with_near_text({"concepts": ["animals"],
                    "distance": max_distance})
    .with_limit(10)
    .with_additional(["distance"])
    .do()
)

print(json.dumps(response, indent=2))
```
